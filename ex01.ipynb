```python
n=12
q=linspace(0,1,n)
d=np.arange(0,n).reshape(-1,1)
x=np.linspace(0,1,1025)
y=x**d
#plt.plot(x,transpose(y));


def runge(x):
    return(1/(1+(10*x-5)**2))

M=q**d
M.shape

c=linalg.solve(transpose(M),runge(q))
c.shape

p=c.dot(y)


plot(x,runge(x))
plot(q,runge(q),"ro")
plot(x,p);
```

```python
linalg.cond(M)
```
